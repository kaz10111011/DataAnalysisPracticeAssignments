## Hypothesis testing

```python
cl.Regression(*de.define_variables(pa.df))
```



I compared the relationship between happiness and various variables : GDP per capita, social support, health expectancy, freedom, generosity. 
- According to the p value above, happiness showed statistically significant positive correlations with social support, health, and freedom. In contrast, generosity did not show a significant correlation with happiness, using a 5% significance level.
- Happiness appears to increase more steeply with Social Support in high-happiness groups than in low-happiness groups. However, the relationship was statistically significant in both groups.

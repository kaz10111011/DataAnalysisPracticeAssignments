### Figure 1

Use `data-fig1.npy` for this figure.

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as plt

a   = np.load( 'data-fig1.npy' )
plt.figure()
plt.plot(a, color = "black", marker ='o' )
plt.xlabel("Time")
plt.ylabel("Value")
plt.show()
```

___

### Figure 2

Use `data-fig2.npy` for this figure.

```python
a   = np.load( 'data-fig2.npy' )
plt.figure()
plt.plot(a[:, 0 ], color = "r")
plt.plot(a[:, 1 ], color = "g")
plt.plot(a[:, 2 ], color = "b")
plt.xlabel("X value") 
plt.ylabel("Y value") 
plt.show()
```

___

### Figure 3

Use `data-fig3.npy` for this figure.

```python
a   = np.load( 'data-fig3.npy' )
plt.figure()
plt.scatter(a[:, 0], a[:, 1], color = "b", s = 200, alpha = 0.4)
plt.xlabel("X value") 
plt.ylabel("Y value") 
plt.show()
```

___

### Figure 4

Use `data-fig4.npy` for this figure.

```python
a   = np.load( 'data-fig4.npy' )
plt.figure()
plt.bar([ 'A', 'B', 'C', 'D', 'E' ],a, color = "0.7")
plt.xlabel("Category") 
plt.ylabel("Value") 
plt.show()
```

___

### Figure 5

Use `data-fig5.npy` for this figure.

*Hint*: Use 2D visualization and the "RdBu" colormap.

```python
a   = np.load( 'data-fig5.npy' )
plt.figure()
plt.imshow(
    a,
    aspect = 'auto',
    cmap = 'RdBu',
    origin = 'lower',
    extent = [0, 41, 0, 120 ],
    vmin = -1.00,
    vmax = 1.00
)
plt.colorbar()
plt.show()
```
